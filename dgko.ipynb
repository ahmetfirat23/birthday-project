```python
import sqlite3 as sql
import csv
from datetime import datetime
import locale
```

Use to display Turkish characters

```python
locale.setlocale(locale.LC_ALL,"Turkish_Turkey.1254")
```

Initialize the database

```python
db = sql.connect("bd.db")
cs = db.cursor()
```

Open csv file 
Check for repitition
Add rows into a temporary list

```python
with open("dgko.csv","r",encoding="utf-8") as file:
    reader = csv.DictReader(file)
    templ=[]
    name=[]
    for row in reader:
        if row["Ad Soyad"] not in name:
            templ.append(tuple(row.values()))
            name.append(row["Ad Soyad"])
            
        
```

Create the database
Add rows to database in a format of name, month.day
(Using database just for practing
A better solution could be found by using sort method with lambda function)

```python
cs.execute("""CREATE TABLE IF NOT EXISTS bd ("ad_soyad", tarih)""")
for date, name, date in templ:
    date = datetime.strptime(date, "%d.%m.%Y")
    cs.execute("""INSERT INTO bd Values(?,?)""",(name.strip().title(), datetime.strftime(date,"%m.%d")))
db.commit()
```

Open a text file for the output
Print the name and month with the format of day month_name

```python
with open("dgko.txt","w",encoding="utf-8") as text:
    cs.execute("""SELECT * FROM bd ORDER BY tarih""")
    list_len = len(cs.fetchall())
    cs.execute("""SELECT * FROM bd ORDER BY tarih""")
    for i in range(list_len):
        tempt = cs.fetchone()        
        print("{}: {:>{align}} {}"
              .format(tempt[0],tempt[1][3:],datetime.strftime(datetime.strptime(tempt[1], "%m.%d"),"%B"),align=30-len(tempt[0]))
             ,file=text)
db.close()        
```
